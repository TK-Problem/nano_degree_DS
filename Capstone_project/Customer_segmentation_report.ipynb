## 2. Main customer architypes

First, data dimesionality was reduced using pricinpal component analysis (PCA) and then it was fitted into **15** clusters using unsupervised learning Kmeans algorithm.

<img src="images/clusters.png" alt="drawing" width="800"/>

### 2.1 Upper class car owners

Cluster **#2** groups people who drive upper class cars (based on engine, max speed, brand, ect.). To illustrate this difference, figures bellow show most most common car segment in the microcell and most common engine size in the microcell.


<center> KBA05_MAXSEG </center>| <center> KBA05_MOTOR </center>
:-------------------------:|:-------------------------:
![](images/car_segment.png)  |  ![](images/engine_size.png)

People in this cluster also tend to live in more densly populated areas, hence advertisement in cities could be recommended. 


### 2.2 Elderly budget car drivers 

Cluster **#13** shows older people (age range 50-60) who prefer to have budget cars, e.g. open or ford. Compared to general population, customers ten to have lower mobility (moving patterns).

<center> General population </center>| <center> Customers dataset </center>
:-------------------------:|:-------------------------:
![](images/mobility_azdias.png)  |  ![](images/mobility_customers.png)

The older person is the less likely he/she is to move arround (~45% of custoemrs have low or very low mobility). Not suprisingly, mail ordersing services looks interesting for these people (ordering stuff from catalog isntead of driving to the supermarket).

### 2.3 Wealthy and environmentally conscious

Cluster **#8** is under-represented in `customers` DataFrame. People in this cluster are wealthy households, have positive affinity towards online services, who drive cars without pre-owners (new cars).

While customers are on average wealthier than general population and are more enviromentally concious, other features of this clusters makes customers dataset underrepresented. People in this cluster would not be interested in wasting paper on mail-ordering catalogs, hence are unlikely to order such service. This hypothesis is suported by comparing onine affinity (tendancies to shop online) between general population and customers dataset.

<center> General population </center>| <center> Customers dataset </center>
:-------------------------:|:-------------------------:
![](images/online_affinity_a.png)  |  ![](images/online_affinity_c.png)

On average customers have lower affinity to online shoping.


### 2.4 Unemployed people from rural areas

Cluster **#14** is also under-represented in `customers` DataFrame. This cluster describes people who drive budget cars, live in rural areas (lower density zones). 

<img src="images/pop_density.png" alt="drawing" width="800"/>

From figure above, it's clear that customers ten to avoid extreemly densly populated as well as rural (low density) areas.

## tl;dr

* Information about potential customer's car class is the best predictor for identifying future customers.

* Customer architype to target:

    * **Upper class car owners**,
    * **Elderly budget car drivers**.
    
* Customer architype to avoid sending promotions:

    * **Wealthy and environmentally conscious**,
    * **Unemployed people from rural areas**.
